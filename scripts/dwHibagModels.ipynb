```python
import requests
from bs4 import BeautifulSoup
from collections import defaultdict
import os
import subprocess
from subprocess import PIPE
from random import randint
from time import sleep
import csv
```

```python
page = open('/media/adiamb/HDD1/getHLAModels/scripts/urls')
soup = BeautifulSoup(page.read())
```

```python
dwUrls=soup.find_all('a', href=True)
```

```python
urlDict = defaultdict(list)
for u in dwUrls:
    eth, url_=u.text, u['href']
    if url_:
        if 'HLA4-hg19' in url_ :
            if eth == "Multi-ethnic models":
                eth = 'MultiEthnic'
            urlDict[eth].append(url_.strip('.'))
        else:
            continue

    
```

```python
urlDict
```

```python
baseUrl = "https://hibag.s3.amazonaws.com"
basePath = "/media/adiamb/HDD1/getHLAModels"
for eth, uList in urlDict.items():
    makePath = os.path.join(basePath, eth)
    print(makePath)
    if os.path.exists(makePath):
        for url in uList:
            fUrl = baseUrl + url
            ModelName = os.path.basename(fUrl)
            fullPath = os.path.join(makePath, ModelName)
            if not os.path.exists(fullPath):
                wgetComd = 'wget {} -P {}'.format(fUrl, makePath)
                print(wgetComd)
                cmd = subprocess.Popen(wgetComd, shell=True, stdout=PIPE, stderr=PIPE)
                stdout, stderr = cmd.communicate()
                #if not stderr:
                tSleep = randint(10,50)
                print('SLEEPING {}'.format(tSleep))
                sleep(tSleep)
            else:
                print('{} EXISTS< SKIPPING DWLD'.format(fullPath))
    else:
        os.mkdir(makePath)

```

```python
## write out the models for future use
with open('scripts/modelPathUrls.txt', 'w') as outFile:
    writer=csv.writer(outFile, delimiter=' ')
    header=['ETH', 'URL']
    writer.writerow(header)
    for eth, uList in urlDict.items():
        for url in uList:
            rows = [eth, baseUrl+url]
            writer.writerow(rows)
            
        
        
```
